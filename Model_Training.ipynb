```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import re
import string
```

```python
# Load true news
data_true = pd.read_csv('True.csv')

# Load fake news
data_fake = pd.read_csv('Fake.csv')
```

```python
data_fake=data_fake.head(21000)
```

```python
data_true=data_true.head(21000)
```

```python
data_fake["class"]=0
data_true["class"]=1
```

```python
data_fake_manual_testing=data_fake.tail(10)
for i in range(20999,20991,-1):
    data_fake.drop([i],axis=0,inplace=True)

data_true_manual_testing=data_true.tail(10)
for i in range(20999,20991,-1):
    data_true.drop([i],axis=0,inplace=True)
```

```python
data_fake.shape,data_true.shape
```

```python
data_fake_manual_testing['class']=0
data_true_manual_testing['class']=1
```

```python
data_merge=pd.concat([data_fake,data_true],axis=0)
data_merge.head(10)
```

```python
data= data_merge.drop(['title','subject','date'],axis=1)
```

```python
data.isnull().sum()
```

```python
data=data.sample(frac=1)
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+/www\.S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]'% re.escape(string.punctuation),'',text)
    text=re.sub('\n','',text)
    text=re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
data
```

```python
x=data["text"]
y=data["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer( 
    stop_words='english',
    max_df=0.9,
    min_df=2,
    ngram_range=(1, 2))
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(penalty="l2")
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
print(f"test accuracy:{LR.score(xv_test,y_test)}")
```

```python
print(classification_report(y_test, pred_lr))
```

```python
def output_lable(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Not A Fake News"

def manual_testing(news):
    testings_news={"text":[news]}
    new_def_test=pd.DataFrame(testings_news)
    new_def_test["text"]=new_def_test["text"].apply(wordopt)
    new_x_test=new_def_test["text"]
    new_xv_test= vectorization.transform(new_x_test)
    pred_LR=LR.predict(new_xv_test)

    return print(f"\nLR Prediction: {output_lable(pred_LR[0])}")
```

```python
news = str(input("Enter the news text to check: "))
manual_testing(news)
```
